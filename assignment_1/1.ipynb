## Part A — Tensor Basics

Create:

- a random tensor A of shape (3, 4)
- a random tensor B of shape (4, 2)

Then:

- 1.Perform matrix multiplication
- 2.Print result shape

Find:

- mean
- max
- standard deviation

```python
import torch

# Creating two tensors a and b
a=torch.rand(3,4)
b=torch.rand(4,2)

# Matrix multiplication
c=a@b
print(c)
print(c.shape)

# Mean, Maximum,Standard Deviation
print(torch.mean(c))
print(torch.max(c))
print(torch.std(c))
```

## Part B — Autograd
Create a tensor:

- values: [1.0, 2.0, 3.0]
- gradients enabled

Compute:

$$y = \sum (x^2 + 2x)$$

Use autograd to compute gradients.


```python
# Creating tensor
x=torch.tensor([1.0,2.0,3.0],requires_grad=True)
y=torch.sum(x**2+2*x)

#Computing Gradients
y.backward()
print(x.grad)
```

## Part C — Mini Challenge
Without using built-in loss functions:

- 1.Create prediction tensor
- 2.Create target tensor
- 3.Manually compute MSE loss

Formula:

$$MSE = \frac{1}{n} \sum (y_{pred} - y_{true})^2$$

```python
# Predciction Tensor
y_pred=torch.tensor([1., 2., 3., 4., 5.])

# Target Tensor
y_true=torch.tensor([1.5, 2.5, 3.5,4.5,5.5])

#MSE
n=len(y_pred)
sum=torch.sum((y_pred-y_true)**2)
mse=sum/n
print(mse)
```

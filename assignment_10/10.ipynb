## Part B — Build CNN

Architecture:

- Conv2d(1,8,kernel_size=3)
- ReLU
- MaxPool2d(2)

- Conv2d(8,16,kernel_size=3)
- ReLU
- MaxPool2d(2)

- Flatten
- Linear(64,3)

```python
import torch 
import torch.nn as nn

class CNN_Model(nn.Module):

    def __init__(self):
        super().__init__()

        self.conv_1=nn.Conv2d(
            in_channels=1,
            out_channels=8,
            kernel_size=3
        )
        self.relu_1=nn.ReLU()
        self.pool_1 = nn.MaxPool2d(kernel_size=2)

        self.conv_2=nn.Conv2d(
            in_channels=8,
            out_channels=16,
            kernel_size=3
        )
        self.relu_2=nn.ReLU()
        self.pool_2=nn.MaxPool2d(kernel_size=2)

        self.flatten=nn.Flatten()
        self.linear=nn.Linear(64,3)
 
    def forward(self,x):

        x=self.conv_1(x)
        x=self.relu_1(x)
        x=self.pool_1(x)

        x=self.conv_2(x)
        x=self.relu_2(x)
        x=self.pool_2(x)

        x=self.flatten(x)
        x=self.linear(x)

        return x

model=CNN_Model()
```

# Part A — Reshape and Flatten

Create a tensor containing numbers from 1 to 24.

Shape it into: (2, 3, 4)

Then:

- Print original tensor
- Print shape
- Flatten it into a 1D tensor
- Reshape flattened tensor into shape: (4, 6)

```python
import torch

# Create tensor from 1 to 24
x=torch.arange(1,25)

# Reshaping into (2,3,4)
y=x.reshape(2,3,4)
print(y)
print(y.shape)

#Flattening into 1D Tensor
z=y.reshape(-1)
print(z)

#Reshape into (4,6)
m=z.reshape(4,6)
print(m)
```

## Part B — Indexing and Slicing

Using the (2,3,4) tensor:

 1. Print first block
 2. Print second row of first block
 3. Print last column from all rows and all blocks
 4. Print only even numbers
- Do NOT use loops for even numbers.

```python
# (2,3,4) tensor
print(y)

# First Block
print(y[0])

#Second Row of first block
print(y[0,1])

# Last column form all rows and all blocks
print(y[:,:,-1])

# Print only even numbers
print(y[y%2==0])
```

## Part C — Broadcasting

Create:

- A shape -> (3,4)
- B shape -> (4,)

Perform broadcasting addition:

- A + B

Then explain:

- why broadcasting worked
- how PyTorch matched dimensions

```python
# Create two tensors A and B of shape (3,4) and (4,)
a=torch.rand(3,4)
b=torch.rand(4,)
print(a,b)

# Perform A+B
add = a+b
print(add)
print(add.shape)
```

### Explanation

For Broadcating , PyTorch compares dimensions.
Two dimensions are compatible if:

- they are equal
- OR one of them is 1

Here:

- A = (3,4)
- B = (4,)

PyTorch aligns from the right side:

- (3,4)
- (  4)

So it interprets B as:

- (1,4)

Now compare:

- 4 == 4 ✔
- 1 can expand to 3 ✔

Final output:

- (3,4)
